Acessar API do IBGE Para extrair os dado da cultura "Soja"

```python
import requests
import json

# Lista de códigos para substituir na URL
codigos = [11, 12, 13, 14, 15, 16, 17, 21, 22, 23, 24, 25, 26, 27, 28, 29, 31, 32, 33, 35, 41, 42, 43, 50, 51, 52, 53]

# URL base
url_base = 'https://servicodados.ibge.gov.br/api/v3/agregados/5457/periodos/-6/variaveis/8331|216|214|112?localidades=N6[N3[{}]]&classificacao=782[40124]'

# Lista para armazenar os dados
dados_ibge = []

# Itera sobre os códigos
for codigo in codigos:
    # Monta a URL substituindo o código
    url = url_base.format(codigo)

    # Realiza a requisição GET
    response = requests.get(url)

    # Verifica se a requisição foi bem-sucedida (código 200)
    if response.status_code == 200:
        # Converte a resposta para JSON
        data = response.json()
        
        # Adiciona os dados à lista
        dados_ibge.extend(data)
    else:
        # Imprime uma mensagem de erro caso a requisição falhe
        print(f"Falha na requisição para código {codigo}. Código de status: {response.status_code}")

# Salva o JSON em um arquivo
with open("C:/Users/anaph/OneDrive/Área de Trabalho/MBA/TCC/Coleta de Dados/IBGE/producao_soja.json", "w", encoding="utf-8") as json_file:
    json.dump(dados_ibge, json_file, ensure_ascii=False, indent=4)

print("JSON salvo com sucesso em 'producao_soja.json'")
```

Ajustando arquivo json para xlsx e organizando extrutura

```python
import pandas as pd
import json

# Caminho para o arquivo JSON
caminho_json = 'C:/Users/anaph/OneDrive/Área de Trabalho/MBA/TCC/Coleta de Dados/IBGE/producao_soja.json'

# Lê o arquivo JSON
with open(caminho_json, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Lista para armazenar os dados processados
processed_data = []

# Itera sobre os resultados do JSON
for resultado in data:
    # Obtém informações comuns
    variavel = resultado.get('variavel', '')
    unidade = resultado.get('unidade', '')
    
    # Itera sobre os resultados
    for resultado_info in resultado.get('resultados', []):
        # Obtém a categoria, se disponível
        categoria = resultado_info.get('classificacoes', [{}])[0].get('categoria', {}).get('40124', '')
        
        # Itera sobre as séries
        for serie in resultado_info.get('series', []):
            localidade_id = serie.get('localidade', {}).get('id', '')
            localidade_nome = serie.get('localidade', {}).get('nome', '')

            # Itera sobre os anos e valores
            for ano, valor in serie.get('serie', {}).items():
                processed_data.append([localidade_id, variavel, unidade, categoria, localidade_nome, ano, valor])

# Cria um DataFrame do pandas com os dados processados
df = pd.DataFrame(processed_data, columns=['CdIbge', 'Variavel', 'Unidade', 'Cultura', 'Localidade', 'Ano', 'Valor'])

# Salva o DataFrame em um arquivo Excel
df.to_excel("C:/Users/anaph/OneDrive/Área de Trabalho/MBA/TCC/Coleta de Dados/IBGE/producao_soja.xlsx", index=False)
```
